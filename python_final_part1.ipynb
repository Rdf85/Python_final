```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
data = pd.read_csv('airbnb_listings.csv')
```

```python
print(data)
```

```python
data.head()
```

```python
data.info()
```

```python
# Droping rows with missing values in 'review_scores_rating' and 'host_is_superhost'
cleaned_data = data.dropna(subset=['review_scores_rating', 'host_is_superhost'])
```

```python
# Converting 'host_is_superhost' to a categorical variable for better visualization
cleaned_data['host_is_superhost'] = cleaned_data['host_is_superhost'].astype('category')
```

```python
# Exploring trends

# Trend 1: Boxplot of 'review_scores_rating' by 'host_is_superhost'
plt.figure(figsize=(10, 6))
sns.boxplot(x='host_is_superhost', y='review_scores_rating', data=cleaned_data)
plt.title('Distribution of Review Scores by Superhost Status')
plt.xlabel('Host is Superhost')
plt.ylabel('Review Scores Rating')
plt.show()
```

```python
# Statistical analysis
superhost_mean = cleaned_data[cleaned_data['host_is_superhost'] == 't']['review_scores_rating'].mean()
non_superhost_mean = cleaned_data[cleaned_data['host_is_superhost'] == 'f']['review_scores_rating'].mean()

print(f"Mean Review Scores for Superhosts: {superhost_mean:.2f}")
print(f"Mean Review Scores for Non-Superhosts: {non_superhost_mean:.2f}")
```

```python
# Compared to non-superhosts (4.56), superhosts have a mean review score of 4.86, which is noticeably higher,Compared to non-superhosts (4.56),
# This suggests that, in comparison to listings without superhosts, listings with superhosts typically have higher review scores.
# Superhosting seems to be linked to better review scores, which could help with booking success and reputation building.

```

```python
# Trend 2: Distribution of 'price' by 'property_type'
plt.figure(figsize=(12, 8))
sns.stripplot(x='property_type', y='price', data=data, jitter=True, palette='Set2')
plt.title('Distribution of Prices by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# It seems like that anything which is a entire accommodation has a bigger price in comparison with shared acomodation.
# That is logical because it has more space, the larger the space the higher the price.
```

```python
# Identify numeric columns
numeric_columns = cleaned_data.select_dtypes(include=['float64', 'int64']).columns

# Trend 3: Correlation Heatmap (excluding non-numeric columns)
correlation_matrix = cleaned_data[numeric_columns].corr()
plt.figure(figsize=(22, 13))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```

```python
# If zoom in you can see a relation in between for example reviews score communication and review score location.
# or review score location and review score vallue.
```

```python
# Trend 4: Distribution of 'accommodates' across different 'room_types'
plt.figure(figsize=(10, 6))
sns.violinplot(x='room_type', y='accommodates', data=data, palette='viridis')
plt.title('Distribution of Accommodates by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Accommodates')
plt.show()
```

```python
# This violin plot shows that Entire home or apartment have the tendency to have higher prices.
# In the other hand accommodations with less square metters are cheaper with hotel rooms being slightly more expensive.
```

```python
# Trend 5: Histogram of 'number_of_reviews' with different 'instant_bookable' status
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='number_of_reviews', hue='instant_bookable', element='poly', kde=True, palette='muted')
plt.title('Distribution of Number of Reviews by Instant Bookable Status')
plt.xlabel('Number of Reviews')
plt.ylabel('Density')
plt.show()
```

```python
# The his histogram indicates that around 20% of the instant bookable will give a review.
```
